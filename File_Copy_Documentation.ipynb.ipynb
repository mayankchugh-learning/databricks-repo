# File Copy Documentation: bronze_layer to silver_layer

Documentation for copying `2_bronze_pyspark.ipynb` from bronze_layer to silver_layer directory

## Overview

This notebook documents the process of copying a notebook file from one directory to another within the Databricks workspace.

**Task**: Copy `2_bronze_pyspark.ipynb` from the bronze_layer directory to the silver_layer directory

**Operation Type**: Copy (not move) - the original file will remain in bronze_layer

**Source**: `databricks-repo/Databricks/bronze_layer/2_bronze_pyspark.ipynb`

**Destination**: `databricks-repo/Databricks/silver_layer/2_bronze_pyspark.ipynb`

## Step 1: Import Required Libraries

We'll use Python's built-in libraries for file operations:
* `os` - for path operations and directory checks
* `shutil` - for high-level file operations including copying with metadata preservation

```python
import os
import shutil

print("✓ Libraries imported successfully")
```

## Step 2: Define Source and Destination Paths

Define the complete paths for:
* Source file location (bronze_layer)
* Destination directory (silver_layer)
* Destination file path

```python
# Define source file path
source_file = "/Workspace/Users/mayankchugh.learning@gmail.com/databricks-repo/Databricks/bronze_layer/2_bronze_pyspark.ipynb"

# Define destination directory
dest_dir = "/Workspace/Users/mayankchugh.learning@gmail.com/databricks-repo/Databricks/silver_layer"

# Define destination file path
dest_file = os.path.join(dest_dir, "2_bronze_pyspark.ipynb")

print(f"Source: {source_file}")
print(f"Destination directory: {dest_dir}")
print(f"Destination file: {dest_file}")
```

## Step 3: Check if Destination Directory Exists

Before copying, we need to ensure the silver_layer directory exists. If it doesn't, we'll create it.

```python
# Check if destination directory exists
if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)
    print(f"✓ Created directory: {dest_dir}")
else:
    print(f"✓ Directory already exists: {dest_dir}")

# Verify the directory exists
if os.path.isdir(dest_dir):
    print(f"✓ Destination directory verified")
```

## Step 4: Copy the File

Use `shutil.copy2()` to copy the file with metadata preservation (timestamps, permissions).

```python
# Copy the file from bronze_layer to silver_layer
shutil.copy2(source_file, dest_file)

print(f"✓ File copied successfully")
print(f"  From: {source_file}")
print(f"  To: {dest_file}")
```

## Step 5: Verify the Copy Operation

Confirm that:
* Both source and destination files exist
* File sizes match
* Original file remains in bronze_layer

```python
# Verify both files exist
source_exists = os.path.exists(source_file)
dest_exists = os.path.exists(dest_file)

print("Verification Results:")
print(f"  Source file exists: {source_exists}")
print(f"  Destination file exists: {dest_exists}")

if source_exists and dest_exists:
    # Compare file sizes
    source_size = os.path.getsize(source_file)
    dest_size = os.path.getsize(dest_file)
    
    print(f"\nFile Sizes:")
    print(f"  Source: {source_size:,} bytes")
    print(f"  Destination: {dest_size:,} bytes")
    
    if source_size == dest_size:
        print(f"\n✓ Copy verified successfully - file sizes match")
    else:
        print(f"\n⚠ Warning: File sizes do not match")
else:
    print(f"\n✗ Verification failed - one or both files not found")
```
