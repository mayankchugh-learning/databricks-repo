## **JSON File [Mapping Files]**

```python
import pandas as pd
import ast

# Read the file as Python dict (since it uses Python syntax)
with open('/Workspace/Users/mayankchugh.learning@gmail.com/databricks-repo/Databricks/silver_layer/mapping.json', 'r') as f:
    content = f.read()
    # Parse as Python literal (handles sets and converts to lists)
    data = ast.literal_eval(content)
    # Convert sets to lists for DataFrame compatibility
    data = {k: list(v) if isinstance(v, set) else v for k, v in data.items()}

df = pd.DataFrame(data)

df.head()
```
